#Zomato-Restaurant-Market-Analysis/

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
 df=pd.read_csv("/Users/haiderali/Downloads/Zomato data .csv")
df
```

```python
def handleRate(value):
    value=str(value).split('/')
    value=value[0];
    return float(value)
df['rate']=df['rate'].apply(handleRate)
print(df.head())
```

```python
#TYPES OF RESTURANTS
```

```python
df.head()
```

```python
sns.countplot(x=df['listed_in(type)'])
plt.xlabel("Type of Resturant")
plt.title("Most Common Restaurant Type")
```

Insight:
Dining restaurants dominate the dataset, indicating that dine-in experiences remain the most common business model among listed establishments. Delivery and Cafes occupy a smaller market share.

```python
group_data=df.groupby('listed_in(type)')['votes'].sum()
result=pd.DataFrame({'votes' : group_data})
plt.plot(result,c='green',marker='o')
plt.xlabel("Type of Resturant",c="green",size=20)
plt.ylabel("Votes",c="red",size=20)
plt.title("Votes by Restaurant Type")
```

Insight:
Dining restaurants have accumulated the highest number of customer votes, suggesting greater customer engagement and popularity compared to other restaurant formats.

```python
plt.hist(df['rate'],bins=5)
plt.title("Distribution of Ratings")
plt.show()
```

Insight:
The majority of restaurants have ratings clustered between 3.5 and 4.2, which suggests that most listed establishments maintain moderate to good customer satisfaction levels.

```python
sns.countplot(x=df['approx_cost(for two people)'])

plt.title("Average order Spending by couples")
```

```python
sns.countplot(x=df['online_order'])
plt.title("Online Order Availability")
plt.show()
```

Insight:
A significant portion of restaurants provide online ordering services, reflecting the increasing importance of digital food delivery channels in the restaurant industry.

```python
pivot_table=df.pivot_table(index='listed_in(type)',columns='online_order',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table,annot =True,fmt='d')
plt.title("Heatmap")
plt.xlabel("Online order")
plt.ylabel("listed_in(type)")
plt.show()
```

Insight:
A moderate positive relationship is observed between restaurant votes and ratings, implying that highly rated restaurants often attract greater customer attention and feedback.

```python
sns.countplot(x=df['book_table'])
plt.title("Table Booking Availability")
plt.show()
```

Insight:
Most restaurants do not offer table booking, indicating that reservation-based dining is less common and limited to selected premium establishments.

```python
sns.boxplot(x='listed_in(type)', y='rate', data=df)
plt.xticks(rotation=45)
plt.title("Restaurant Type vs Ratings")
plt.show()
```

Insight:
Dining and Buffet restaurants exhibit comparatively stable and higher ratings, whereas quick delivery outlets show wider variation in customer satisfaction.

```python
# Key Findings
```

1. Dining restaurants are the most dominant and most voted category.
2. Online ordering significantly improves restaurant engagement.
3. Table booking is mostly available in premium restaurants.
4. Mid-range restaurants capture the largest customer segment.
5. Higher ratings are generally associated with higher customer votes.
6. Customer convenience features directly impact restaurant popularity.

```python
# Business Recommendations
```

1. Restaurants should invest in online ordering systems to improve customer reach.
2. Mid-range pricing appears to attract the largest customer base, making it a sustainable pricing model.
3. Premium restaurants can leverage table booking to improve customer experience.
4. Restaurants with low ratings but high visibility should focus on service quality improvement.
5. Food delivery focused outlets should optimize customer reviews to increase trust and votes.

```python
# Conclusion
```

This analysis of the Zomato restaurant dataset highlights that customer convenience, service model, and pricing strategy play a crucial role in determining restaurant popularity. Dining establishments remain the most preferred category, while digital features such as online ordering enhance customer interaction. The study provides actionable insights for restaurant owners to improve visibility, engagement, and customer satisfaction.
